###  Изучаем стобец id

Проверим наличие дубликатов в столбце id.

```python
dupl_id = df[df.duplicated(subset='id', keep = False)]
dupl_id
```

В столбце id обнужились 5 пар дублированных значений. Между дублями разница в основном в названиях вакансий (незначительная), дате публикации и навыках. В остальных столбцах данные совпадают. Можем сделать предположение, что это повторные публикации вакансий с внесением некоторых уточнений.

Удалим дубликаты вакансий, оставив только те, что были опубликованы позже.

```python
#отсортируюем дубликаты по дате публикации (сначала ранние)
dupl_id_sort = dupl_id.sort_values(by=['id', 'published_at'], ascending=[True, True])
#создадим датафрейм, куда войдут строки с дублями, которые были опубликованы первыми
min_date_dupl = dupl_id_sort.drop_duplicates(subset='id', keep='first')
min_date_dupl
```

```python
#удалим из датафрейма строки, которые есть в датафрейме с отсортированными дублями
df = df[~(df['id'].isin(min_date_dupl['id']) & df['published_at'].isin(min_date_dupl['published_at']))]
#проверим наличие дубликатов id
df[df.duplicated(subset='id', keep = False)]
```

Нам удалось удалить 5 дублей.

### Изучаем стобец area

```python
#проверим качество записи значений
#sorted(df['area'].unique())
```

```python
#выведем количество уникальных значений
df['area'].value_counts()
```

Большинство офисов нанимающих компаний находится в Москве. Затем по количеству идет Санкт-Петербург, далее крупные города России.  

Во многих городах размещено по 1-2 вакансии. Их можно объединить.
Так же в дальнейшем интересно будет посмотерть на разницу в зарплатах между столичными регионами и остальными. Проверим теорию, влияет ли близость к столице на ЗП. 

Поэтому создадим столбец с распределением всех городов на Москву, Санкт-Петербург, Московскую область, а остальные города запишем в  'Регионы'.

```python
#создаем функцию
def region(row):
    if 'Москва' in row:
        return 'Москва'
    elif 'Санкт-Петербург' in row:
        return 'Санкт-Петербург'
    elif 'Бал' in row or 'Дзержинский' in row or 'Дмит' in row or 'Дуб' in row or 'Кор' in row \
    or 'Красног' in row or 'Лух' in row or 'Мыт' in row or 'Один' in row or 'Под' in row \
    or 'Пуш' in row or 'Ступ' in row or 'Хим' in row or 'Щер' in row:
        return 'Московская область'
    else: 
        return 'Регионы'

#создаем новый столец и применяем функцию    
df['area_region'] = df['area'].apply(region)
```

### Изучаем стобцы с указанием зарплаты: salary_from, salary_to, salary_bin

Посмотрим на значения в столбце salary_bin

Количество значений 'ЗП не указана'в salary_bin совпадает с количеством пропусков в salary_to.   

Ранее мы заметили, что значение в столбце "ЗП не указана" иногда стоит в тех строках, где в salary_from есть значения. Выведем количество таких строк.

```python
filtered_df_to = df[(df['salary_to'].notna()) & (df['salary_bin'] == 'ЗП не указана')]
filtered_df_from = df[(df['salary_from'].notna()) & (df['salary_bin'] == 'ЗП не указана')]
print('Строк со значениями в salary_to, но ЗП не указана:', filtered_df_to.shape[0])
print('Строк со значениями в salary_from, но ЗП не указана:', filtered_df_from.shape[0])
```

Можем сделать вывод, что значения для столбца salary_bin ориентированы на столбец  salary_to, а заполненные значения в salary_from остались незамеченными.  

Исправим это, заменив значения в salary_bin в соответвии со значениями в столбце salary_from, только для тех строк, где в salary_to стоят пропуски.


```python
#создаем условие для вставки значений
condition = (df['salary_from'].notna()) & (df['salary_to'].isna()) & (df['salary_bin'] == 'ЗП не указана')

#создаем функцию для вставки значений
def sal_bin(row):
    if 0 < row < 100000:
        return 'Меньше 100 тысяч'
    elif 100000 <= row < 200000:
        return 'От 100 тысяч до 200 тысяч'
    elif 200000 <= row < 300000:
        return 'От 200 тысяч до 300 тысяч'
    elif row >= 300000:
        return 'Больше 300 тысяч'
    else:
        return 'ЗП не указана'
    
#применяем функцию с условием
df.loc[condition, 'salary_bin'] = df.loc[condition, 'salary_from'].apply(sal_bin)
```

Пропуски удалены.

После замены неявных дублей в строках появились явные дубликаты, если было перечисление похожих навыков. Избавимся от них.

```python
#создаем функцию для удаления дубликатов в строке навыков
def remove_duplicates(row):
    #разбиваем строку на список 
    sk_list = row.split(', ')
    #преобразуем список в множество для удаления дублей, затем обратно в список
    unique_sk_list = list(set(sk_list))
    #сортируем
    unique_sk_list.sort()
    #объединяем обратно в строку через запятую
    return ', '.join(unique_sk_list)


#применяем функцию ко всем строкам столбца с очищенными навыками
df['cleaned_skills'] = df['cleaned_skills'].apply(remove_duplicates)
```

**Приступим к разделению навыков на хард и софт**

Создадим вручную список из софт скиллов, на основе данных из столбца со всеми скиллами.

```python
soft = ['коммуникативные навыки',
        'проактивность', 
        'работа в команде', 
        'системное мышление', 
        'критическое мышление', 
        'стратегическое мышление', 
        'логическое мышление', 
        'стрессоустойчивость', 
        'внимание к деталям', 
        'добросовестность', 
        'организаторские навыки', 
        'математический склад ума', 
        'точность и внимательность к деталям', 
        'грамотная речь', 
        'ответственность', 
        'наставничество', 
        'умение работать в коллективе', 
        'умение принимать решения']
```

Теперь запустим проверку и создадим новые столбцы с хард и софт скиллами.

```python
#создаем пустые списки для софт и хард скиллов
soft_skills_list = []
hard_skills_list = []

#перебираем все строки в столбце cleaned_skills
for skills in df['cleaned_skills']:
    #разделяем навыки по запятой и убираем лишние пробелы
    skill_list = [skill.strip() for skill in skills.split(',')]
    
    #временные списки для текущих софт и хард скиллов
    current_soft_skills = []
    current_hard_skills = []
    
    #проверяем каждый навык
    for skill in skill_list:
        if skill in soft:
            current_soft_skills.append(skill)  
        else:
            current_hard_skills.append(skill)  

    #объединяем навыки обратно в строки
    soft_skills_list.append(', '.join(current_soft_skills))
    hard_skills_list.append(', '.join(current_hard_skills))

#добавляем новые столбцы в датафрейм
df['soft_skills'] = soft_skills_list
df['hard_skills'] = hard_skills_list
```

### Определяем какие навыки требуются чаще  - хард или софт, для каждого грейда двух профессий.

```python
#создаем фунукцию
def skills_by_experience(dff):

    #создаем временный датафрейм
    temp = dff.copy()
    
    #добавляем столбцы с количеством всех скиллов, хард и софт
    temp['all_skills_count'] = temp['all_skills'].apply(lambda x: len(x.split(',')) if pd.notna(x) else 0)
    temp['hard_skills_count'] = temp['hard_skills'].apply(lambda x: len(x.split(',')) if pd.notna(x) else 0)
    temp['soft_skills_count'] = temp['soft_skills'].apply(lambda x: len(x.split(',')) if pd.notna(x) else 0)

    #группируем по грейдам и считаем среднее количество скиллов
    skills_by_experience = temp.groupby('experience').agg({
        'all_skills_count': 'mean',
        'hard_skills_count': 'mean',
        'soft_skills_count': 'mean'
    }).reset_index()

    #подготовка данных для построения графика
    skills_melted = skills_by_experience.melt(id_vars='experience', 
                                              value_vars=['all_skills_count', 'hard_skills_count', 'soft_skills_count'], 
                                              var_name='skill_type', value_name='mean_skills')
    #сортируем данные
    skills_melted_sorted = skills_melted.sort_values(by=['mean_skills', 'experience'], ascending=[False, False])
    
    return skills_melted_sorted
```

```python
plt.figure(figsize=(12, 6))

#строим график DA
ax1=plt.subplot(1, 2, 1)
sns.barplot(data=skills_by_experience(da_df), x='experience', y='mean_skills', hue='skill_type', palette='viridis', ax=ax1)

#добавляем подписи и заголовок
plt.title('Среднее количество скиллов по грейдам DA')
plt.xlabel('Грейды')
plt.ylabel('Среднее количество скиллов')
plt.legend(title='Тип скиллов')

#строим график SA
ax2=plt.subplot(1, 2, 2, sharey=ax1)
sns.barplot(data=skills_by_experience(sa_df), x='experience', y='mean_skills', hue='skill_type', palette='viridis', ax=ax2)

#добавляем подписи и заголовок
plt.title('Среднее количество скиллов по грейдам SA')
plt.xlabel('Грейды')
plt.ylabel('Среднее количество скиллов')
plt.legend(title='Тип скиллов')

#отображаем графики
plt.tight_layout()
plt.show()
```

Для аналитиков данных распределение ожидаемое и логичное - больше всего навыков указывают в требованиях к вакансиям грейда senior, меньше всего для специалистов без опыта - junior.

А для профессии Системный аналитик данных распределение другое - больше всего навыков указывают для грейда junior, далее идет грейд senior, и меньше всего навыков указывают для middle.

Преобладание хард скиллов заментно для обоих профессий во всех грейдах, в среднем в вакансиях указывают 1 мягкий навык. 

### Определяем самые важные хард и софт скиллы, для каждого грейда двух профессий.

Построим общею функцию для построения графиков

```python
def top_skills (dff, exp):

    plt.figure(figsize=(8, 6))

    #Первый график софт скиллс
    #готовим данные 
    ax1 = plt.subplot(2, 1, 1)
    top_soft = (
        (dff.loc[(dff['experience'] == exp), 'soft_skills']
        .str.split(',')
        .explode()
        .str.lstrip()
        .loc[lambda x: x != ""]  
        .value_counts(normalize=True) * 100)  
        .head()
        .sort_values(ascending=False)  
        .reset_index()  
    )

    #переименовываем столбцы
    top_soft.columns = ['soft_skills', 'percentage']

    #строим график 
    sns.barplot(
        data=top_soft, 
        y='soft_skills', 
        x='percentage', 
        palette='viridis',
        ax=ax1
    )
 
    #добавляем заголовок и подписи
    plt.xlabel('Процент вакансий')
    plt.title('Топ-5 софт скиллс')
    plt.ylabel('')


    #Второй график хард скиллс
    ax2 = plt.subplot(2, 1, 2)
    top_hard = (
        (dff.loc[(dff['experience'] == exp), 'hard_skills']
        .str.split(',')
        .explode()
        .str.lstrip()
        .loc[lambda x: (x != "") & (x != "навыки не указаны")] 
        .value_counts(normalize=True) * 100)  
        .head()
        .sort_values(ascending=False)  
        .reset_index()  
    )

    top_hard.columns = ['hard_skills', 'percentage']

    sns.barplot(
        data=top_hard, 
        y='hard_skills', 
        x='percentage', 
        palette='viridis',
        ax=ax2
    )

    plt.xlabel('Процент вакансий')
    plt.title('Топ-5 хард скиллс')
    plt.ylabel('')
    
    #выставляем растояние между графиками и вызываем график
    plt.subplots_adjust(hspace=.5)
    plt.show()
```

### Определяем зависимость уровня заработной платы и региона места работы

Создадим общую функцию для двух профессий.

```python
def area_salary(dff):

    #уберем строки без указания ЗП
    dff_filtered = dff[dff['salary_bin'] != 'ЗП не указана']
    
    #создадим таблицу сопряженности для регионов и зарплатных категорий
    salary_area_ct = pd.crosstab(dff_filtered['area_region'], dff_filtered['salary_bin'], normalize='index') * 100

    #преобразуем данные для seaborn
    salary_area_ct_reset = (
        salary_area_ct.reset_index()
        .melt(id_vars='area_region', var_name='salary_bin', value_name='percentage')
        .sort_values(by=['percentage', 'area_region'], ascending=[False, True])
    )
    
    #строим график
    plt.figure(figsize=(10, 6))
    sns.barplot(x='area_region', y='percentage', hue='salary_bin', data=salary_area_ct_reset, palette='viridis')

    plt.title('Зависимость зарплатных категорий от региона')
    plt.xlabel('Регион')
    plt.ylabel('Процент зарплатных категорий')
    plt.legend(title='Категории зарплат', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
    
```
